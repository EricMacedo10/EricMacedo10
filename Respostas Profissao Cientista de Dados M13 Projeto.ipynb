# 4 - Plote um gráfico de barras onde temos a média de descontos por categoria.

```python
# Seu código aqui

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = ("C:\\Users\\ericm\\Downloads\\MODULO7_PROJETOFINAL_BASE_SUPERMERCADO - MODULO7_PROJETOFINAL_BASE_SUPERMERCADO (1).csv.csv")
df_full = pd.read_csv(file_path)

media_descontos = df_full.groupby('Categoria')['Desconto'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,7))
bars = plt.bar(media_descontos.index, media_descontos, color=sns.color_palette("pastel", len(media_descontos)), edgecolor='black')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f"{int(bar.get_height())}%", ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title("Média de descontos por Categoria", fontsize=14, fontweight='bold')
plt.xlabel("Categoria", fontsize=12)
plt.ylabel("Média de descontos (%)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.tight_layout()
plt.show()
```

# 5 - Plote um gráfico de mapa interativo agrupando os dados por categoria, marca e trazendo a média de desconto.

```python
# Seu código aqui

import pandas as pd
import plotly.express as px

file_path = ("C:\\Users\\ericm\\Downloads\\MODULO7_PROJETOFINAL_BASE_SUPERMERCADO - MODULO7_PROJETOFINAL_BASE_SUPERMERCADO (1).csv.csv")

try:
    df_full = pd.read_csv(file_path, delimiter=',', on_bad_lines='skip')
    print("Colunas carregadas:", df_full.columns)
    
    if 'Desconto' not in df_full.columns:
        print("Erro: A coluna 'Desconto' não foi encontrada. Verifique o nome real da coluna no arquivo CSV.")
    else:  
        df_full['Desconto'] = pd.to_numeric(df_full['Desconto'], errors='coerce')
        
        df_filtrado = df_full[df_full['Desconto'] > 0]
    
        media_desconto = df_filtrado.groupby(['Categoria', 'Marca'])['Desconto'].mean().reset_index()

        fig = px.treemap(media_desconto, path=['Categoria', 'Marca'], values='Desconto', color='Desconto', color_continuous_scale=px.colors.sequential.Viridis, title = 'Média de desconto por Categora e Marca')
        fig.update_layout(title_font_size=20, title_font_family="Arial", title_font_color="darkblue", title_x=0.5, margin=dict(t=50, l=25, r=25, b=25))

        output_html_path = "C:\\Users\\ericm\\Downloads\\Mapa_Interativo.html"
        fig.write_html(output_html_path)

        print(f"Gráfico interativo salvo em: {output_html_path}")
        
except FileNotFoundError:
    print("Erro: O arquivo não foi encontrado. Verifique o caminho e tente novamente.")
except pd.errors.ParserError as e:
    print(f"Erro ao processar o arquivo CSV: {e}")
except Exception as e: 
    print(f"Ocorreu um erro inesperado: {e}") 
```
